# Matplotlib
Pythonのグラフ描画ライブラリである

計算結果の視覚化を目的としている

```python
"""読み込み
"""
from matplotlib import pyplot as plt
import matplotlib
import numpy as np

plt.rcParams["font.family"] = "IPAexGothic"
```

```python
"""線グラフ1
y=x
"""
x=[1,2,3,4,5,6,7]
y=[1,2,3,4,5,6,7]
plt.plot(x, y)
plt.show()
```

```python
"""線グラフ2
y=x^2
"""
x=np.linspace(0, 100, 10000) 
y=x**2
plt.plot(x, y)
plt.show()
```

```python
"""線グラフ3
y=x^2
"""
x=np.linspace(0, 100, 10000) 
y=x**2
plt.title('2次関数')#タイトル
plt.xlabel('X')#X軸名
plt.ylabel('Y')#Y軸名
plt.plot(x, y,label='2次曲線')#凡例名
plt.legend()#凡例の表示
plt.show()#グラフの表示
```

```python
"""線グラフ4
y=x^2
"""
x=np.linspace(0, 100, 10000) 
y1=x**2
y2=(x+10)**2
plt.title('2次関数')#タイトル
plt.xlabel('X')#X軸名
plt.ylabel('Y')#Y軸名
plt.plot(x, y1,label='x^2')#凡例名
plt.plot(x, y2,label='(x+10)^2')#凡例名
plt.legend()#凡例の表示
plt.show()#グラフの表示
```

```python
"""散布図1
"""
x=np.random.randn(100)
y=np.random.randn(100)
plt.title('散布図')#タイトル
plt.xlabel('X')#X軸名
plt.ylabel('Y')#Y軸名
plt.scatter(x, y,label="散布図")#凡例名
plt.legend()#凡例の表示
plt.show()#グラフの表示
```

```python
"""散布図2
"""
x=np.random.randn(100)
y1=np.random.randn(100)
y2=np.random.normal(3,1,100)   
plt.title('散布図')#タイトル
plt.xlabel('X')#X軸名
plt.ylabel('Y')#Y軸名
plt.scatter(x, y1,label="散布図1")#凡例名
plt.scatter(x, y2,label="散布図2")#凡例名
plt.legend()#凡例の表示
plt.show()#グラフの表示
```
